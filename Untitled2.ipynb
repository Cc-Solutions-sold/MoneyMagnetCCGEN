```python
import PySimpleGUI as sg
import sqlite3
import os

db_file = 'approval_tracker.db'

def init_db():
    conn = sqlite3.connect(db_file)
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS approvals
                 (id INTEGER PRIMARY KEY AUTOINCREMENT,
                  numbers TEXT,
                  approval_status TEXT,
                  timestamp DATETIME DEFAULT CURRENT_TIMESTAMP)''')
    conn.commit()
    conn.close()

def save_to_db(numbers, approval_status):
    conn = sqlite3.connect(db_file)
    c = conn.cursor()
    c.execute("INSERT INTO approvals (numbers, approval_status) VALUES (?, ?)",
              (numbers, approval_status))
    conn.commit()
    conn.close()

def fetch_db():
    conn = sqlite3.connect(db_file)
    c = conn.cursor()
    c.execute("SELECT * FROM approvals")
    rows = c.fetchall()
    conn.close()
    return [f"{row[1]} - {row[2]} on {row[3]}" for row in rows]

init_db()

def home_page():
    layout = [
        [sg.Text('Enter 12 numbers separated by commas:', size=(40,1))],
        [sg.InputText(key='numbers_input', size=(40,1))],
        [sg.Button('Submit'), sg.Button('View Database')]
    ]
    return sg.Window('Home', layout, finalize=True)

def results_page(numbers, approval_status):
    layout = [
        [sg.Text(f'Numbers: {numbers}', size=(40,1))],
        [sg.Text(f'Status: {approval_status}', size=(40,1))],
        [sg.Button('Save to Database'), sg.Button('Back to Home')]
    ]
    return sg.Window('Results', layout, finalize=True)

def database_page():
    history = fetch_db()
    layout = [
        [sg.Text('Approval History:', size=(40,1))],
        [sg.Listbox(values=history, size=(60, 10), key='database_list')],
        [sg.Button('Export'), sg.Button('Clear History'), sg.Button('Back to Home')]
    ]
    return sg.Window('Database', layout, finalize=True)

home_win, results_win, db_win = home_page(), None, None

while True:
    window, event, values = sg.read_all_windows()

    if event == sg.WIN_CLOSED:
        window.close()
        if window == home_win:
            break
        if window == results_win:
            results_win = None
        if window == db_win:
            db_win = None

    if event == 'Submit' and window == home_win:
        numbers = values['numbers_input']
        approval_status = 'Approved'  # Placeholder for API call
        home_win.hide()
        results_win = results_page(numbers, approval_status)

    if event == 'Save to Database' and window == results_win:
        save_to_db(numbers, approval_status)
        sg.popup('Results saved!')
        results_win.close()
        results_win = None
        home_win.un_hide()

    if event == 'View Database' and window == home_win:
        home_win.hide()
        db_win = database_page()

    if event == 'Back to Home' and (window == results_win or window == db_win):
        window.close()
        home_win.un_hide()

    if event == 'Clear History' and window == db_win:
        conn = sqlite3.connect(db_file)
        c = conn.cursor()
        c.execute("DELETE FROM approvals")
        conn.commit()
        conn.close()
        sg.popup('History cleared!')
        db_win.close()
        db_win = database_page()

    if event == 'Export' and window == db_win:
        with open('approval_history.txt', 'w') as f:
            for entry in fetch_db():
                f.write(entry + '\n')
        sg.popup('History exported to approval_history.txt!')

home_win.close()
```
